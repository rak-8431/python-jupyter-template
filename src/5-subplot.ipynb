### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for different plots
x = np.linspace(1, 10, 10)
y = x ** 2  # Line chart data

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]  # Bar chart data

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]  # Scatter plot data

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]  # Pie chart data

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot - Line chart
axs[0, 0].plot(x, y, color="blue", marker="o", linestyle="-")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("Y")

# Second subplot - Bar chart
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Third subplot - Scatter plot
axs[1, 0].scatter(scatter_x, scatter_y, color="orange", marker="o", s=100)
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X")
axs[1, 0].set_ylabel("Y")

# Fourth subplot - Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'orange'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
